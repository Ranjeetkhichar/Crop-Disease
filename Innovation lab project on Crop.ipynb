# Importing all required libraries

```python
from keras.models import Sequential
from keras_preprocessing.image import ImageDataGenerator
from keras import regularizers, optimizers
import pandas as pd
import numpy as np
from keras.layers import Dense, Activation, Flatten, Dropout, BatchNormalization 
from keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras import layers
import tensorflow as tf
```

# generating train and test dataframe

```python
traindf=pd.read_csv(r'/Users/ranjeetkhinchar/Downloads/data/train/train/train.csv') 

testdf=pd.read_csv(r'/Users/ranjeetkhinchar/Downloads/data/test/test.csv') 

datagen=ImageDataGenerator(rescale=1./255.)
print(traindf.head(5))
print(testdf.head(5))
```

# shuffling datas

```python
traindf = traindf.sample(frac=1).reset_index(drop=True) 
traindf.head(10)
```

# changing Dtype from object to float

```python
traindf.dtypes,traindf.info()
```

```python
traindf["healthy_wheat"] = pd.to_numeric(traindf["healthy_wheat"]) 
traindf["leaf_rust"] = pd.to_numeric(traindf["leaf_rust"]) 
traindf["stem_rust"] = pd.to_numeric(traindf["stem_rust"]) 
traindf.dtypes,traindf.info()
```

# shape of training datas

```python
traindf.shape
```

# Generating train generator iterator

```python
columns=["healthy_wheat","leaf_rust","stem_rust"]
train_generator=datagen.flow_from_dataframe(
dataframe=traindf[:876],
directory=r'/Users/ranjeetkhinchar/Downloads/data/train/train/All training image/',
x_col="Image",
y_col=columns, 
batch_size=8,

seed=42, 
shuffle=True, 
class_mode="raw", 
target_size=(32,32) 
)
```

# Generating validate generator iterator

```python
valid_generator=datagen.flow_from_dataframe(
dataframe=traindf[400:562],
directory=r"/Users/ranjeetkhinchar/Downloads/data/train/train/All training image/",
x_col="Image",
y_col=columns,
batch_size=8,
seed=42,
shuffle=True,
class_mode="raw",
target_size=(32,32)
)
```

# building model

```python
# from sklearn.ensemble import RandomForestClassifier as Random_forest
# model=Random_forest()
model = Sequential() 
model.add(
    Conv2D(32, (3, 3),padding='same',input_shape=(32,32,3),activation='relu') 
)
model.add( 
    Conv2D(64,(3,3),activation='relu')
) 
model.add(
    MaxPooling2D(pool_size=(2, 2)) 
)
model.add( 
    Dropout(0.25)
) 
model.add(
    Conv2D(128, (3, 3),padding='same',activation='relu') 
)
model.add( 
    MaxPooling2D(pool_size=(2, 2))
) 
model.add(
    Dropout(0.25)
)
model.add(Flatten()) 
model.add(
    Dropout(0.5)
)
model.add( 
    Dense(3,activation='softmax')        
)
```

# test data generating

```python
test_generator=datagen.flow_from_dataframe(
dataframe=testdf,
directory=r"/Users/ranjeetkhinchar/Downloads/data/test/test",
x_col="Image",
batch_size=1, #always for testing
shuffle=False,
seed=42,
class_mode=None, 
target_size=(32,32)
)
```

```python
model.summary()
```

# Total layers used in model

```python
len(model.layers)
```

# compiling model

```python
model.compile( 
    optimizer='Adam', 
    loss="binary_crossentropy", 
    metrics=["accuracy"]
)
```

# fitting model

```python
history=model.fit( 
    train_generator,
    batch_size = 8,
    steps_per_epoch=562//8, #total images in training dataset//batch_size 
    validation_data= valid_generator, #its validation sample
    validation_steps=104//8, #total validation sample
    epochs=10,
)
```

# Plotting model loses

```python
from matplotlib import pyplot as plt 
plt.plot(history.history['loss']) 
plt.plot(history.history['val_loss']) 
plt.xlim(-1,11)
plt.ylim(0,1)
plt.title("Model train vs validation loss") 
plt.ylabel("Loss")
plt.xlabel("Epochs") 
plt.legend(["train","validations"]) 
plt.show()
```

# Plotting model Accuracy


```python
from matplotlib import pyplot as plt 
plt.plot(history.history['accuracy']) 
plt.plot(history.history['val_accuracy']) 
plt.xlim(-1,11)
plt.ylim(0,1)
plt.title("Model train vs validation accuracy") 
plt.ylabel("accuracy")
plt.xlabel("Epochs") 
plt.legend(["train","validations"])
plt.show()
```
